```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math
```

```python
r0 = 2.
k = 1
L = 10
omega = 1
npoints = 1024
xmax = 60
```

```python
Phi0 = np.exp(- (np.linspace(-xmax/2, xmax/2, npoints))**2/r0**2)
Phi0hat = np.fft.fftshift(np.fft.fft(Phi0))
freq = 2 * np.pi * (np.arange(Phi0hat.shape[0]) /Phi0hat.shape[0]-1/2)/((xmax)/npoints)
Phithat = Phi0hat * np.exp(- np.complex(0, 1) * freq**2 *L/(2*k))
Phit = np.fft.ifft(np.fft.ifftshift(Phithat))
```

```python
plt.plot(np.linspace(-xmax/2, xmax/2, npoints), Phi0)
plt.plot(np.linspace(-xmax/2, xmax/2, npoints), np.absolute(Phit)**2)
```

```python
Rt = r0 * np.sqrt(1+4*L**2/(k**2*r0**4))
mod_theorical = r0/Rt*np.exp(-2*np.linspace(-xmax/2, xmax/2, npoints)**2/Rt**2)
```

```python
plt.plot(np.linspace(-xmax/2, xmax/2, npoints), np.absolute(Phit)**2)
plt.plot(np.linspace(-xmax/2, xmax/2, npoints), mod_theorical)
```

```python
zc = 1
h=1
assert zc==h
#np.random.seed(1)
xc=4
sigma=1
mus = np.zeros((npoints, int(math.ceil(L/zc))))
for i in range(int(math.ceil(L/zc))):
    X=np.linspace(-xmax/2, xmax/2, npoints)
    W=np.random.randn(npoints )
    R=sigma**2*np.exp(-X**2/xc**2)
    f=np.fft.fft(np.fft.fftshift(R))
    F=np.fft.ifft( np.sqrt(f)*np.fft.fft(W) )
    mus[:, i] = np.real(F)
    
def D(Phi, h, k):
    Phihat = np.fft.fftshift(np.fft.fft(Phi))
    Phihat *= np.exp(- np.complex(0, 1) * freq**2 *h/(2*k))
    Phit = np.fft.ifft(np.fft.ifftshift(Phihat))
    return Phit

def N(Phi, h, k, mu):
    return Phi * np.exp(1j * k/2 * h * mu)
    
Phi = np.exp(- (np.linspace(-xmax/2, xmax/2, npoints))**2/r0**2)
def ssf(Phi, mus, k, h):
    for i in range(int(math.ceil(L/zc))):
        #Linear half step:
        Phi = D(Phi, h/2, k)

        #Non linear step:
        Phi = N(Phi, h, k, mus[:, i])

        #Linear half step:
        Phi = D(Phi, h/2, k)
    return Phi
```

```python
plt.plot(np.linspace(-xmax/2, xmax/2, npoints), np.absolute(ssf(Phi, mus, k, h))**2)
```
